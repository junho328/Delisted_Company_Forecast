# 1. Paired t-test

```python
delisted = np.log(delisted)
going = np.log(going)
```

```python
# 신뢰수준 95%
for i in range(1,27):
  deli = np.array(delisted[f'X{i}'].tolist())
  go = np.array(going[f'X{i}'].tolist())

  stat, p_val = stats.ttest_rel(deli,go)
  decision = 'O'
  if p_val > 0.05:
    decision = 'X'
  
  print(f'X{i}','  statistic:', stat, '  p-value:', p_val, '  inclusion:', decision)
```

```python
# 신뢰수준 99%
for i in range(1,27):
  deli = np.array(delisted[f'X{i}'].tolist())
  go = np.array(going[f'X{i}'].tolist())

  stat, p_val = stats.ttest_rel(deli,go)
  decision = 'O'
  if p_val > 0.01:
    decision = 'X'
  
  print(f'X{i}','  statistic:', stat, '  p-value:', p_val, '  inclusion:', decision)
```

# 2. Logistic Regression

```python
# y label 추가
delisted['Y'] = 0
going['Y'] = 1
```

```python
delisted.head(2)
```

```python
# 데이터 합치기
df = pd.concat([delisted,going],ignore_index=True)
df.head(2)
```

```python
# 영향 주는 변수만 남기고 제거
df = df[['X2','X5','X6','X7','X13','X14','X15','X22','Y']]
df
```

```python
# 독립 변수
features = df[['X2','X5','X6','X7','X13','X14','X15','X22']]

# 종속 변수
label = df['Y']
```

```python
# 학습 데이터 평가 데이터 분리
from sklearn.model_selection import train_test_split

train_features, test_features, train_labels, test_labels = train_test_split(features, label, train_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(train_features, train_labels)
```

logistic regression은 실행할 때 regularation을 사용하기 때문에 무조건 정규화를 해야함

```python
# StandardScaler 적용 시 
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
train_features = scaler.fit_transform(train_features)
test_features = scaler.transform(test_features)

model = LogisticRegression()
model.fit(train_features, train_labels)

print(model.score(test_features, test_labels))
print(model.coef_)
```

```python
# 여러 번 반복. 평균 성능
accuracy_list = []
coef_list = [[0],[0],[0],[0],[0],[0],[0],[0]]
coefs = [2,5,6,7,13,14,15,22]
for _ in range(1000):

  train_features, test_features, train_labels, test_labels = train_test_split(features, label, train_size=0.2, stratify=label)
  scaler = StandardScaler()
  train_features = scaler.fit_transform(train_features)
  test_features = scaler.transform(test_features)

  model = LogisticRegression()
  model.fit(train_features, train_labels)

  accuracy = model.score(test_features,test_labels)

  accuracy_list.append(accuracy)

  for i in range(8):
    coef_list[i].append(model.coef_[0][i])

print(f"1000번 시행 평균 성능 : {np.mean(accuracy_list)}")
for i in range(8):
  print(f"X{coefs[i]} 평균 계수:{np.sum(coef_list[i])/1000}")
```
